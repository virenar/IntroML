```python
param_kenerl = ['poly', 'rbf', 'linear', 'sigmoid']
param_degree = np.arange(1, 10, 1)
param_C = np.arange(0.1, 2, 0.01)
param_gamma = np.arange(1, 50, 1)
```

```python
from itertools import product
```

```python
param_search = list(product(param_kenerl, param_C))
```

```python
def _fit_model(model, X, y):
    # TODO: Fit the model.
    model.fit(X,y)
    
    # TODO: Make predictions. Store them in the variable y_pred.
    y_pred = model.predict(X)
    
    # TODO: Calculate the accuracy and assign it to the variable acc.
    acc = accuracy_score(y,y_pred)
    return acc
```

```python
# TODO: Create the model and assign it to the variable model.
# Find the right parameters for this model to achieve 100% accuracy on the dataset.
accs = []
for i in param_search:
    k, c = i
    if k == 'poly':
        for d in param_degree:
            model = SVC(C=c, kernel=k, degree=d, gamma='auto')
            acc = _fit_model(model, X, y)
            if acc == 1:
                print(i, d)
            accs.append(acc)
    elif k == 'rbf':
        for g in param_gamma:
            model = SVC(C=c, kernel=k, gamma=g)
            acc = _fit_model(model, X, y)
            if acc == 1:
                print(i,g)
            accs.append(acc)

```
